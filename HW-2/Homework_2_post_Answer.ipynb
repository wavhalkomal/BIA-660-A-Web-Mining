Choose one of your favorite book at goodreads.com (e.g. for the book "The Midninght Library", the URL is: https://www.goodreads.com/book/show/52578297-the-midnight-library?from_choice=true)

- Q1. Write a function to scrape all **reviews on the first page**, including, 
    - **reviewer's name** (see (1) in Figure)
    - **number of reviews** (see (2) in Figure)
    - **number of followers** (see (3) in Figure)
    - **rating** (see (4) in Figure)
    - **date** (see (5) in Figure)
    - **review content** (see (6) in Figure. For each review text, need to get the **complete text**, i.e., need to expand the `more` button. Only text is needed, pictures are not needed. Hint: take a close look at the content of the html file.)
    - If a field, e.g. score, is missing, use `None` to indicate it. 
- `Function Input`: book page URL
- `Function Output`: save all reviews as a DataFrame of columns (`reviewer, number, follower, rating, date, review, like`). E.g., for the given URL, you can get 30 reviews.

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions 
import requests
import time
from bs4 import BeautifulSoup  
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions 


def getReviews(page_url):
        

        
    driver = webdriver.Chrome()
    
    # send a request
    driver.get(page_url)

    # enter your code
    driver.maximize_window()         
    
    pagecheck=False
    while pagecheck!=True:
        if "wait" in driver.title:
            driver.refresh()
        elif "The Midnight Library by Matt Haig | Goodreads" in driver.title:
            print('title found now move to next step')
            pagecheck=True
            
    time.sleep(10)
    
    cnt = driver.find_element(By.XPATH,'//div/div[6]/div[1]/span').text
    cnt = cnt.split('of')
    cnt = cnt[0]
    cnt = cnt.split('-')
    cnt = int(cnt[1]) + 1
 

    ################################################################################
    ##reviewer's name (see (1) in Figure)
    reviewer = []
    pagecheck=False
    while pagecheck!=True:
        try:
            for i in range(1,cnt):
                xpath =  '(//div/article/div//div[@class="ReviewerProfile__name"])[' + str(i) + ']'
                reviewer_data = driver.find_element(By.XPATH,xpath)
                reviewer.append(reviewer_data.text)
                
            pagecheck = True
        except:
            print("--------Finding reviewer-----------")
            pagecheck=False
    #print("--------reviewer found-----------",reviewer)
 
    ################################################################################
    ## number of reviews (see (2) in Figure)
    
    ################################################################################
    ################################################################################
    ## number of followers (see (3) in Figure)
    
    number = []
    follower = []
    pagecheck=False
    while pagecheck!=True: 
        for i in range(1,cnt):
            try:
                xpath_ = '(//*[@id="ReviewsSection"]//div[@class="ReviewerProfile__meta"])['  + str(i) + ']/span[1]'
                num = driver.find_element(By.XPATH, xpath_).text 
                xpath__ = '(//*[@id="ReviewsSection"]//div[@class="ReviewerProfile__meta"])['  + str(i) + ']//span/span'
                
                foll = driver.find_element(By.XPATH, xpath__).text 
                
                number.append(num ) 
                follower.append( foll )
                 
                pagecheck = True
            except:
                print("--------Finding number of reviews -----------")
                print("--------Finding number of follower -----------")
                pagecheck=False

    #print("--------number of reviews found-----------",number)
    #print("--------number of follower found-----------",follower)
   
    ################################################################################
    ################################################################################
    ## date (see (5) in Figure)
    date = []
    pagecheck=False
    while pagecheck!=True:
        for i in range(1,cnt):
            try:
                xpath = '(//*[@class="ReviewCard__row"]/span/a)['  + str(i) + ']'
                date_ = driver.find_element(By.XPATH, xpath).text  
                date_ = date_.replace('First published' ,'').strip()
                date.append(date_)
                pagecheck = True
            except:
                print("--------Finding date -----------")
                pagecheck=False

    #print("--------date found-----------",date)
        
    
    ################################################################################
    ################################################################################
    ## rating (see (4) in Figure)
    rating = []
    pagecheck=False
    while pagecheck!=True:
        for i in range(1,cnt):
            try:
                xpath = '(//*[@class="ReviewsList"]//*[contains(@aria-label,"Rating")])['  + str(i) + ']'
                
                valll = driver.find_element(By.XPATH, xpath).get_attribute("aria-label")
                print(valll)
                rating.append(valll  ) 
            except:pass
                #print("--------Finding number of rating -----------")
                #pagecheck=False
                
        pagecheck = True
    print("--------number of rating found-----------",rating)


    ################################################################################
    ################################################################################
    ## review content (see (6) in Figure. For each review text, need to get the complete text, i.e., need to expand the more button. Only text is needed, pictures are not needed. Hint: take a close look at the content of the html file.)
    ## If a field, e.g. score, is missing, use None to indicate it.
    
    review = []
    pagecheck=False
    while pagecheck!=True:
        
        for i in range(1,cnt):
            try: 
                    xxpath = '(//div/button[@class="Button Button--tertiary Button--medium"]/span[2][@class="Button__labelItem"])['   + str(i) + ']'
                    driver.find_element(By.XPATH, xxpath).click()  
                    print('clicked ', i) 
                    
            except:pass
                #print("--------Finding review -----------")
                #pagecheck=False
        
        pagecheck = True
        
    time.sleep(10)
    print('now get the review')
    pagecheck=False
    while pagecheck!=True:

        for i in range(1,cnt): 
            try:
                xpath = '(//*/article/section/section[2]/section/div/div[1]/span)[' + str(i) + ']'
                print('found ', i )
                review_ = driver.find_element(By.XPATH, xpath).text
                print('found ', review_)
                
                review_ = review_.strip()
                review.append(review_)
                pagecheck = True
                
            except:
                print("--------Finding review -----------")
                pagecheck=False
                
    print("--------review found-----------",review)    

     
    # Create a dictionary with the lists
    data = {
        'reviewer': reviewer,
        'number': number,
        'follower': follower,
        'rating': rating,
        'date': date,
        'review': review
    }
    
    # Create the DataFrame
    df = pd.DataFrame(data)
    
    # Display the DataFrame
    print(df)
    
    


page_url = 'https://www.goodreads.com/book/show/52578297-the-midnight-library?from_choice=true'
getReviews(page_url)

    
```
